```python
def get_dates():
    date_list = []

    start_date = date.today() - relativedelta(years=1) + relativedelta(days=1)
    end_date = date.today()
    asofdate = start_date

    while(asofdate <= end_date):
        date_list.append(
            str(asofdate.year) + 
            "-" + 
            str(asofdate.month).zfill(2) + 
            "-" + 
            str(asofdate.day).zfill(2)
        )
        asofdate = asofdate + relativedelta(days=1)

    return date_list



def get_news(ticker):
    print("Downloading {} news...".format(ticker))
    length_list = np.zeros(len(date_list))  

    company_frame = pd.DataFrame(
        columns=[
            'category',
            'datetime',
            'headline',
            'id',
            'image',
            'related',
            'source',
            'summary',
            'url'
        ]
    )

    for i in trange(len(date_list)):
        rest_url = (
            'https://finnhub.io/api/v1/company-news?symbol=' + 
            ticker + 
            '&from=' + 
            date_list[i] + 
            '&to=' + 
            date_list[i] + 
            '&token=' + 
            token
        )
        results = requests.get(rest_url).json()

        for k in range(len(results)):
            cur_doc = results[k]
            company_frame = company_frame.append(cur_doc,ignore_index=True)
        time.sleep(1)
    
    company_frame['summary_length'] = company_frame['summary'].apply(len)
    company_frame.to_pickle(
        '/content/drive/MyDrive/abnormal-distribution-project-data/news/' +
        'articles_' + 
        ticker + 
        '.pkl'
    )
```
